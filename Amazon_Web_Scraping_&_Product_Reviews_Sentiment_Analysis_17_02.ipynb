```
# requests - package used to get html page from URL/ URI

import requests
from bs4 import BeautifulSoup as bs
```

```
url = 'https://www.amazon.in/New-Apple-iPhone-12-128GB/product-reviews/B08L5S1NT7/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber=2'
```

```
page = requests.get(url)
```

```
page.content
```

```
# scrap page content using html.parser , xml.parser 
soup = bs(page.content, "html.parser") # source of web page
```

```
soup
```

```
# Prettify() - transform unformatted html page into readable html data
print(soup.prettify())
```

```
# find-all : span tag name & tag-class 'a-profile-name'
names = soup.find_all('span', class_ = 'a-profile-name')
names
```

```
cust_names = []
for i in range(len(names)):
    cust_names.append(names[i].get_text())
```

```
cust_names
```

```
# Find all title content & tag class
title = list(soup.find_all(class_ = 'review-title'))
title
```

```
review_title = []
for i in range(len(title)):
    review_title.append(title[i].get_text())
```

```
review_title
```

```
# To remove '\n' , '\t' & other such Noise. str method strip will be used to elliminate unwanted characters 
review_title[:] = [review_title.strip('\n') for review_title in review_title]
review_title 
```

**Rating**

```
ratings = list(soup.find_all(class_ = 'review-rating'))
ratings
```

```
rating_list = []
for i in range(len(ratings)): 
    rating_list.append(ratings[i].get_text()[0])
```

```
rating_list  
```

```
# typecasting every list values in integer
rating_list[:] = [int(rating_list) for rating_list in rating_list]
rating_list 
```

```
review = list(soup.find_all('span', {"data-hook" : 'review-body'}))
review
```

```
review_body = []
for i in range(len(review)): 
    review_body.append(review[i].get_text())
```

```
review_body
```

```
review_body[:] = [review_body.strip('\n ') for review_body in review_body]
review_body 
```

```
import pandas as pd
```

```
amazon_product_review = pd.DataFrame()

amazon_product_review['Customer Name'] = cust_names[2: ]
amazon_product_review['Review Title'] = review_title[2: ]
amazon_product_review['Review'] = review_body[:]
amazon_product_review['Ratings'] = rating_list[2: ]
```

```
amazon_product_review
```

```
import numpy as np
import seaborn as sns
import nltk
import matplotlib.pyplot as plt
```

```
# Sentiment Intensity Analyzer 
nltk.download("vader_lexicon")
```

```
from nltk.sentiment.vader import SentimentIntensityAnalyzer

siaz = SentimentIntensityAnalyzer()
```

```
siaz.polarity_scores(amazon_product_review.iloc[0, 1]) 
```

```
import pandas as pd
def visualize_sentiment(data):
    sns.heatmap(pd.DataFrame(data).set_index("Sentence").T, annot = True, cmap = "RdPu")
```

```
sentence = amazon_product_review.iloc[1,1]
print(sentence)

plt.figure(figsize = (12,8))

visualize_sentiment({
    "Sentence": [sentence] + sentence.split(), 
    "Sentiment" : [siaz.polarity_scores(sentence)["neu"]] + [siaz.polarity_scores(word)["neu"] for word in sentence.split()]
})
```
